```python
import pandas as pd
file = pd.read_csv("sentiment_analysis_pos_neg.csv")
length = file.index.size
print(length)
AllStocks = []
GreatestRatio=0
IndexOfBestStock=0
for i in range(length):
    IndividualStock = [(file.at[i,"Positives"]), (file.at[i,"Negatives"]), 
                       (file.at[i,"Neutrals"]), (file.at[i,"Undefineds"]), 
                       (file.at[i, "Rating"]), (file.at[i, "Price"])]
    AllStocks += [IndividualStock]
    CurrentRatio = max (IndividualStock[0],IndividualStock[1])/min(IndividualStock[0],IndividualStock[1])
    
    if CurrentRatio>GreatestRatio:
        GreatestRatio = CurrentRatio
        IndexOfBestStock = i

    
    print("i = ", i, " CurrentRatio=", CurrentRatio, " IndexOfBestStocks=", IndexOfBestStock," GreatestRatio=", GreatestRatio)
   

        
print("IndexOFBestStock=", IndexOfBestStock, " GreatestRatio=", GreatestRatio)

print("Top Decision is", IndexOfBestStock, AllStocks[IndexOfBestStock])
if AllStocks[IndexOfBestStock][0] >  AllStocks[IndexOfBestStock][1]:
    print("Buy ^")
elif AllStocks[IndexOfBestStock][1] > AllStocks[IndexOfBestStock][0]:
    print("Sell ^")
```
