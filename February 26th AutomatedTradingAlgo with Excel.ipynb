```python
import pandas as pd
#factor = 100000
factor = 100000
#import the CSV file
dfrate_price = pd.read_csv("IS235Ratings_Prices.csv") #this creates a data frame, so call it dfrate_price
#See how long the file is (fileLength)
lines = dfrate_price.index.size
#start loop from set Rating to Buy or Sell (fileLength).
countBuys = 0
countShorts = 0
countIgnores = 0
for i in range(lines):   
    #set variable to cell of the CSV Rating
    rating = float(dfrate_price.at[i,'Rating'])
    #Set variable to cell of the CSV Price
    price = float(dfrate_price.at[i,'Price'])
    #set variable to get cell to either buy or sell. 
    #if the price is Lower than the (rating*factor), set cell name to buy
    if price <= rating*factor:
        countBuys+=1
        print("buy", "Rating is: ", rating, "Price is: ", price)
    #if the price is higher than the (rating*factor), set cell name to short
    elif price >= rating*factor:
        countShorts+=1
        print("short", "Rating is: ", rating, "Price is: ", price)
    #if the price is the same to the (rating*factor), set cell name to ignore
    elif price == rating*factor:
        countIgnores+=1
        print("ignore", "Rating is: ", rating, "Price is: ", price)
#end of loop
print("The number of buys is: ", countBuys)
print("The number of shorts is: ", countShorts)
print("The number of ignores is: ", countIgnores)
```
