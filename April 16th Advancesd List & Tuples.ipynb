```python
#import pandas as pd
import pandas as pd
#read file using pd
file=pd.read_csv("sentiment_analysis_pos_neg.csv")
#get length of file
length=file.index.size
parentList = []
maxRatio = 0
positionOfMaxRatio = 0
#initialize for loop create variables and link then to elements in file
for i in range (length):
    variableList = [(file.at[i,"Positives"]), (file.at[i,"Negatives"]), (file.at[i,"Neutrals"]), 
                    (file.at[i,"Undefineds"]), (file.at[i, "Rating"]), (file.at[i, "Price"])]
    parentList += [variableList]
    
    ratio = max(variableList[0], variableList[1])/min(variableList[0], variableList[1]) 
    if ratio > maxRatio:
        maxRatio = ratio
        positionOfMaxRatio = i
    
print("the max ratio", maxRatio, positionOfMaxRatio)
```
